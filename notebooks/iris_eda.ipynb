```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")  
```

```python
df = pd.read_csv("../data/raw/iris.csv")  # Adjust path if needed
df.head()
```

```python
df.describe()
df.info()
```

```python
df.isnull().sum()
```

```python
df.hist(figsize=(10, 6))
plt.tight_layout()
plt.show()
```

```python
numeric_df = df.drop("species", axis=1)

plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df, hue="species", height=2.5)
plt.show()
```

# Observations
- High correlation between petal_length and petal_width.
- No missing values.
- The data seems clean and well-separated.
